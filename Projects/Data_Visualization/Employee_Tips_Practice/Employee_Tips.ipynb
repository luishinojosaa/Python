```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('C:/Users/l.hinojosa/Desktop/Python/Projects/Data_Visualization/Employee_Tips_Practice/Datasets/tips.csv') 

pd.options.display.max_columns = None
pd.options.display.max_rows = None

print(df.head())
```

```python
# Scatter Plots

# Scatter plot with tips by day, with table size shown by color, and total bill shown by dot size
plt.scatter(df['day'], df['tip'], c=df['size'], s=df['total_bill'])

# Adding labels
plt.title("Tips by Day")
plt.xlabel('Day')
plt.ylabel('Tips')
plt.colorbar()

plt.show()
```

```python
# Line Chart

# Scatter plot with tips by day, with table size shown by color, and total bill shown by dot size
plt.plot(df['tip'])
plt.plot(df['size'])
# Adding labels
plt.title("Line Plot")
plt.xlabel('Day')
plt.ylabel('Tips')

plt.show()
```

```python
# Bar Chart

# Bar chart with day against tip
plt.bar(df['day'], df['tip'])
 
plt.title("Bar Chart")
 
# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Tip')
 
# Adding the legends
plt.show()
```

```python
# Histogram

# histogram of total_bills
plt.hist(df['total_bill'])
 
plt.title("Histogram")
 
# Adding the legends
plt.show()
```
